```python
X=x_opt
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
X_train=pd.DataFrame(X_train).to_numpy()
```

## Turbo Boost & Vacuum Gauge(psi) Vs Engine Load

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train[:,[7]], y_train)
```

```python
plt.scatter( X_train[:,7], y_train['Engine Load(Absolute)(%)'], color='red')
plt.plot( X_train[:,7],regressor.predict( X_train[:,[7]]))
plt.title('Turbo Boost & Vacuum Gauge(psi) VS Engine Load')
plt.ylabel('Engine Load')
plt.xlabel('Turbo Boost & Vacuum Gauge(psi)')
plt.show()
```
